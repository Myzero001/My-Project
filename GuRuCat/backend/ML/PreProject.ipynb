# **Set Data**



## **Test Data**

```python
import pandas as pd
import numpy as np
df_test = pd.read_csv('/content/testData.csv')
df_test
```

## **Train Data**

```python
df_train = pd.read_csv('/content/MML_data.csv')
df_train
```

# Data prepare

## **Test Data**

**Split Data X and Y** (Test Data)

```python
y_test = df_test['Mood (Label)']
y_test
```

```python
x_test = df_test.drop(["Mood (Label)"],axis='columns')
x_test
```

## **Train Data**

**Split Data X and Y** (Train Data)

```python
y_train = df_train['Mood (Label)']
y_train
```

```python
x_train = df_train.drop(["Mood (Label)"],axis='columns')
x_train
```

# **Train and Build Model**

## **Decision Tree**

### **Train Model**

```python
from sklearn.tree import DecisionTreeClassifier

# Decision Tree
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)

# predict
y_pred = clf.predict(x_test)

y_pred
```

錫錫꿋 overfitting 仙錫錫№仙錫錫錫□링錫錫꿋르 tune model


```python
clf = DecisionTreeClassifier(max_depth=5, min_samples_split=4, random_state=42)
clf.fit(x_train, y_train)

y_pred = clf.predict(x_test)
print(y_pred)
```

## **Random Forest**


```python
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier(n_estimators=10, max_depth=5, random_state=42)
rf_clf.fit(x_train, y_train)

y_pred = rf_clf.predict(x_test)
print(y_pred)
```
